# Wielowymiarowy rozkład normalny

```python
# import matplotlib.pylab as plt
import matplotlib.pyplot as plt
import numpy as np
from mpl_toolkits.mplot3d import Axes3D
import matplotlib.mlab as mlab
from matplotlib import cm
```

# Narysujmy gęstość dwuwymiarowego rozkładu Normalnego

```python
from scipy.stats import multivariate_normal

mu_x = 0
variance_x = 12

mu_y = 0
variance_y = 12
x = np.linspace(-10,10,500)
y = np.linspace(-10,10,500)
        
X = np.arange(-3, 3, 0.001)
Y = np.arange(-3, 3, 0.001)
X, Y = np.meshgrid(x,y)

pos = np.empty(X.shape + (2,))
pos[:, :, 0] = X; pos[:, :, 1] = Y

Z = multivariate_normal([mu_x, mu_y], [[variance_x, 0], [0, variance_y]]).pdf(pos)

fig = plt.figure(figsize=(20,10))
ax = fig.add_subplot(1, 2, 1, projection='3d')
_ = ax.plot_surface(X, Y, Z, linewidth=0.1, cmap=cm.viridis)
plt.show()
```

# Narysujmy kontury gęstości dwuwymiarowego rozkładu Normalnego

```python
fig, ax = plt.subplots(nrows=1, ncols=1, figsize=(5, 5))
ax.set_ylim(-12, 11)
ax.set_xlim(-10, 10)
levels = np.arange(0.0, 0.014, 0.0003)
_ = plt.contour(X, Y, Z, levels)
plt.show()
```

# Wylosujmy próbkę z dwuwymiarowego rozkładu Normalnego

```python
mean = [0, 0]
cov = np.array([[ 0.8 , -0.1], [-0.1,  0.8]])
samples = np.random.multivariate_normal(mean, cov, 600)
w, v = np.linalg.eig(cov)
fig, ax = plt.subplots(nrows=1, ncols=1, figsize=(5, 5))
_ = plt.scatter(samples[:,0], samples[:,1])
plt.show()
```
