```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

from matplotlib.colors import ListedColormap
from help_plot import plot_decision_regions
```

# Random Forests

```python
from sklearn.model_selection import train_test_split
from sklearn import datasets
iris = datasets.load_iris()
list(iris.keys())

X = iris["data"][:, (2, 3)]  # petal length, petal width
y = (iris["target"]).astype(np.int)
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=42)
```

```python
plt.scatter(X_train[:, 0], X_train[:, 1], c=y_train)
plt.show()
```

```python
plt.hist(y_train)
plt.hist(y_test)
plt.show()
```

```python
from sklearn.ensemble import RandomForestClassifier
from help_plot import plot_decision_regions
from sklearn import metrics
```

```python
rfc1 = RandomForestClassifier(n_estimators=500, max_leaf_nodes=16, n_jobs=-1, random_state=42)
rfc1.fit(X_train, y_train)
print(metrics.accuracy_score(y_train, rfc1.predict(X_train)))
print(metrics.accuracy_score(y_test, rfc1.predict(X_test)))
plot_decision_regions(X, y, rfc1)
```

```python
rfc2 = RandomForestClassifier(n_estimators=50, max_leaf_nodes=2, n_jobs=-1, random_state=42)
rfc2.fit(X_train, y_train)
print(metrics.accuracy_score(y_train, rfc2.predict(X_train)))
print(metrics.accuracy_score(y_test, rfc2.predict(X_test)))
plot_decision_regions(X, y, rfc2)
```

```python
rfc3 = RandomForestClassifier(n_estimators=5, max_leaf_nodes=2, n_jobs=-1, random_state=42)
rfc3.fit(X_train, y_train)
print(metrics.accuracy_score(y_train, rfc3.predict(X_train)))
print(metrics.accuracy_score(y_test, rfc3.predict(X_test)))
plot_decision_regions(X, y, rfc3)
```

```python
import pandas as pd
classifiers = [rfc1, rfc2, rfc3]
d = {'precision_score': [metrics.precision_score(y_test, clf.predict(X_test), average='weighted') for clf in classifiers], 
     'recall_score': [metrics.recall_score(y_test, clf.predict(X_test), average='weighted') for clf in classifiers], 
     'f1_score': [metrics.f1_score(y_test, clf.predict(X_test), average='weighted') for clf in classifiers],
     'accuracy_score' : [metrics.accuracy_score(y_test, clf.predict(X_test)) for clf in classifiers]
    }
df = pd.DataFrame(data=d)
df.insert(loc=0, column='Method', value=['rfc1', 'rfc2', 'rfc3'])
df
```
