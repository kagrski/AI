````python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
````

# Decision Trees feature selection

````python
from sklearn import datasets
from sklearn.tree import DecisionTreeClassifier
from sklearn.model_selection import train_test_split

np.random.seed(1)
cancer = datasets.load_breast_cancer()
# print description
# print(cancer.DESCR)
````

````python
# get the data
X = cancer.data
y = cancer.target
````

````python
X_train, X_test, y_train, y_test = train_test_split(X, y, stratify=y, random_state=42)

plt.hist(y_train, alpha=0.5)
plt.hist(y_test, alpha=0.5)
plt.show()
````

````python
tree = DecisionTreeClassifier(max_depth=4, random_state=0)
tree.fit(X_train, y_train)

print("Accuracy on training set: {:.3f}".format(tree.score(X_train, y_train)))
print("Accuracy on test set: {:.3f}".format(tree.score(X_test, y_test)))
````

````python
import matplotlib.pyplot as plt

data = tree.feature_importances_
names = cancer.feature_names

bins = list(range(1,len(data)+1))
plt.figure(figsize=(10,10))
plt.barh(y=bins,width=data,height=0.75)
plt.yticks(list(range(1,len(data)+1)), labels=names)
plt.xlabel("Feature importance")
plt.ylabel("Feature")

plt.show()
````
